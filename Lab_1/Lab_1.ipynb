# Loading/reading image using OpenCv

```python
import cv2
img = cv2.imread('/images/lib.webp') #Load the image data into a NumPy array

import matplotlib.pyplot as plt
plt.imshow(img)
plt.show()
```

# Loading/reading image using PIL

```python
from PIL import Image
img2 = Image.open('/images/d3.webp') #load image into a PIL Image object

import matplotlib.pyplot as plt
import matplotlib.cm as cm
plt.imshow(img2, cmap = cm.Greys_r)
plt.show()
```
